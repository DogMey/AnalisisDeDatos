# Leyendo CSV

## Importando archivos CSV

```python
import pandas as pd


datos = pd.read_csv("superstore_data.csv")
datos.head()
```

### Parametros para el método .read_csv

```python
# sep: Es el separador que se usa en el archivo para separar los datos
# nrows: Indica el número de filas que quiero leer del archivo
# usecols: Indica el numero de columnas de tal manera:

datos_seleccion = pd.read_csv("superstore_data.csv", usecols=["Id", "Year_Birth", "Income"])
datos_seleccion.head()
```

```python
# O sin saber el  nombre pero si la posición
datos_seleccion = pd.read_csv("superstore_data.csv", usecols=[0,1,4])
datos_seleccion
```

## Escribiendo archivos CSV

```python
# De manera predeterminada cuando exportamos pandas nos genera un indice, que muchas veces puede ser datos duplicados si ya lo tenemos en nuestra tabla
datos_seleccion.to_csv("clientes_mercado.csv", index=False)
clientes_mercado = pd.read_csv("clientes_mercado.csv")
clientes_mercado
```
