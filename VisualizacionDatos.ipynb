# Vizualización de datos con Python (Matplotlib, Seabron y Plotly)

```python
import pandas as pd

df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
```

```python
df.head()
```

```python
anos = list(map(str, range(1980, 2014)))
colombia = df.loc['Colombia', anos]

col_dict = {
    "Año": colombia.index.tolist(),
    "Inmigrantes": colombia.values.tolist()
}

datos_col = pd.DataFrame(col_dict)
datos_col.head()
```

## Usando Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.title('Inmigrantes de Colombia')
plt.xlabel('Año')
plt.ylabel('Inmigrantes')
plt.xticks(["1980","1985","1990","1995","2000","2005","2010","2010"])
plt.figure(figsize=(10, 5))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
ax.set_title('Inmigrantes de Colombia a Cánada')
ax.set_xlabel('Año')
ax.set_ylabel('Inmigrantes')
plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(10, 3))


axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_title('Inmigrantes de Colombia a Cánada')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Inmigrantes')
axs[0].grid(True)

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de Inmigrantes')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Inmigrantes')
axs[1].grid(True)

plt.show()
```

```python
datos_col.describe()
```

```python
fig, axs = plt.subplots(2,2,figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Inmigrantes a Cánada')

axs[0,0].plot(df.loc["Colombia", anos])
axs[0,0].set_title('Inmigrantes de Colombia a Cánada')

axs[0,1].plot(df.loc["Brasil", anos])
axs[0,1].set_title('Inmigrantes de Brasil a Cánada')

axs[1,0].plot(df.loc["Argentina", anos])
axs[1,0].set_title('Inmigrantes de Argentina a Cánada')

axs[1,1].plot(df.loc["Perú", anos])
axs[1,1].set_title('Inmigrantes de Perú a Cánada')

ymin = 0
ymax = 7000

for ax in axs.flat:
    ax.set(ylim=(ymin, ymax))

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Año')
    ax.set_ylabel('Inmigrantes')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'],lw=3,color ="r")
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
ax.set_title('Inmigrantes de Colombia a Cánada', fontsize = 18, loc="left")
ax.set_xlabel('Año')
ax.set_ylabel('Inmigrantes')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

fig.savefig("/content/inmigrantesColor.png", transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

```python
sudamerica = df.query("Region == 'América del Sur'")

colores = ["royalblue", "seagreen", "slategray", "orangered", "darkred", "yellowgreen", "brown", "grey", "deeppink", "navy", "teal", "tomato"]

fig, ax = plt.subplots(figsize = (12,3))

ax.barh(sudamerica.index, sudamerica["Total"], color = colores)

ax.set_title("Inmigración de sudamericanos a Canadá", loc="left", fontsize=18)
ax.set_xlabel("País", fontsize=14)
ax.set_ylabel("Inmigrantes", fontsize=14)

ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values(by="Total", ascending=True)

colores = ["royalblue", "seagreen", "slategray", "orangered", "darkred", "yellowgreen", "brown", "grey", "deeppink", "navy", "teal", "tomato"]

fig, ax = plt.subplots(figsize = (12,3))

ax.barh(sudamerica_sorted.index, sudamerica_sorted["Total"], color = colores)

ax.set_title("Inmigración de sudamericanos a Canadá", loc="left", fontsize=18)
ax.set_xlabel("País", fontsize=14)
ax.set_ylabel("Inmigrantes", fontsize=14)

ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colores=[]

for pais in sudamerica_sorted.index:
    if pais == "Colombia":
      colores.append('forestgreen')
    else:
      colores.append('royalblue')

fig, ax = plt.subplots(figsize = (10,3))

ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.tick_params(axis="y", which="both", size=0)

ax.barh(sudamerica_sorted.index, sudamerica_sorted["Total"], color = colores)

ax.set_title("Inmigración de sudamericanos a Canadá", loc="left", fontsize=18)
ax.set_xlabel("País", fontsize=14)

ax.yaxis.set_tick_params(labelsize=10)

for i,j in enumerate(sudamerica_sorted["Total"]):
    ax.text(j, i, str(j), va='center')

plt.show()
```

## Seaborn

```python
import seaborn as sns
```

```python
sns.set_theme()
```

```python
top_10 = df.sort_values("Total", ascending=False).head(10)
top_10
```

```python
ax = sns.barplot(data=top_10, x="Total", y=top_10.index, orient="h")
ax.set(title="Top 10 paises inmigrantes a Canadá", xlabel="Inmigrantes", ylabel="Total")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax = sns.barplot(data=top_10, x="Total", y=top_10.index, orient="h")
ax.set_title("Top 10 paises de inmigración a Canadá", loc="left", fontsize=18)
ax.set_xlabel("Número de inmigrantes", fontsize=16)
ax.set_ylabel("")

plt.show()
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8, 4))
  ax = sns.barplot(data=top_10, x="Total", y=top_10.index, hue=top_10.index, orient="h", palette=palette)
  ax.set_title("Top 10 paises de inmigración a Canadá", loc="left", fontsize=18)
  ax.set_xlabel("Número de inmigrantes", fontsize=16)
  ax.set_ylabel("")

  plt.show()
```

```python
generar_grafico("Blues_r")
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8, 4))
  ax.set(xticklabels=[])
  ax = sns.barplot(data=top_10, x="Total", y=top_10.index, hue=top_10.index, orient="h", palette=palette)
  ax.set_title("Top 10 paises de inmigración a Canadá", loc="left", fontsize=18)
  ax.set_xlabel("Número de inmigrantes", fontsize=16)
  ax.set_ylabel("")
  sns.despine(left=True, bottom=True)

  plt.show()
```

```python
sns.set_theme(style="darkgrid")

generar_grafico("tab10")
```

## Plotly

```python
import plotly.express as px
```

```python
fig = px.line(datos_col, x="Año", y="Inmigrantes", title="Inmigrantes de Colombia a Canadá")
fig.update_traces(line_color="green", line_width=3)
fig.update_layout(width=600, height=400, xaxis={"tickangle":-45}, xaxis_title="Año", yaxis_title="Inmigrantes")
fig.show()
```
